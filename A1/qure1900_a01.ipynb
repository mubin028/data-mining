```python
#1 What is the mean, median, and mode of the data? Comment on the dataâ€™s modality (i.e., bimodal, trimodal, etc.)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Age data
age_data = [13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70]

# Calculate mean, median, mode
mean_age = np.mean(age_data)
mean_age=np.rint(mean_age)
median_age = np.median(age_data)
modes = pd.Series(age_data).mode()

print("Modes:", list(modes))
print("Mean age rounded to nearest int: ", mean_age) 
print("Median age: ", median_age)
print(" ")

#data is bimodal as both the number 25 and 35 appear 4 times each in the data set


#2 calculate 5 num summary and IQR

quartile_1 = np.percentile(age_data, 25)
quartile_3 = np.percentile(age_data, 75)

IQR = quartile_3 - quartile_1
min_age = np.min(age_data)
max_age = np.max(age_data)

print("5 Number Summary:")
print("Min age: ", min_age)
print("Q1: ", quartile_1)
print("Median age: ", median_age)
print("Q3: ", quartile_3)
print("Max age: ", max_age)
print(" ")
print("IQR: ", IQR)
print(" ")

plt.figure(figsize=(10, 5)) 
plt.subplot(1, 2, 1)
plt.hist(age_data, bins=10, edgecolor='k')
plt.xlabel('Age')
plt.ylabel('Count of Frequehcy')
plt.title('Histogram of Age Data')

# Boxplot
plt.subplot(1, 2, 2)
plt.boxplot(age_data, vert=False)
plt.xlabel('Age')
plt.title('Boxplot of Age Data')

# Show the plots
plt.tight_layout()
plt.show()
```
